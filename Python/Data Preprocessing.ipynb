```python
import pandas as pd
import datetime as dt
import numpy as np
from matplotlib import pyplot as plt
pd.options.mode.chained_assignment = None
from sklearn.preprocessing import OneHotEncoder 

old_df = pd.read_csv ("AWCustomers.csv")
print(old_df)

new_df = old_df.drop(['Title','MiddleName','AddressLine2','Suffix'],axis=1)
print(new_df)
```

```python
new_df['CountryRegionName'].unique()
```

```python
new_df['Occupation'].unique()
```

```python
new_df['Education'].value_counts()
```

```python
new_df['Education']=new_df['Education'].map({'Partial High School':1,'High School':2,'Partial College':3,'Bachelors':4,'Graduate Degree':5})
new_df
```

```python
new_df['Occupation'].value_counts()
```

```python
new_df['Occupation']=new_df['Occupation'].map({'Manual':1,'Skilled Manual':2,'Clerical':3,'Management':4,'Professional':5})
new_df
```

```python
new_df.dtypes
```

```python
#Handling Null Values
new_df.isnull()
new_df.isnull().sum()
```

```python
#Normalization
maximum=new_df["YearlyIncome"].max()
minimum=new_df["YearlyIncome"].min()
new_df["YearlyIncomeNormalized"]= (new_df['YearlyIncome']-minimum)/(maximum-minimum)
new_df
```

```python
# Discretization (Convert Birthdate to Age)
from datetime import datetime
today_year=pd.to_datetime('today').year
new_df['BirthDate']=pd.to_datetime(new_df["BirthDate"],format='%d-%m-%Y')
new_df['Age']= today_year - (new_df['BirthDate'].dt.year)
new_df
```

```python
# Standardization
yearInc=new_df['YearlyIncome']
np.array(yearInc)

m=np.mean(yearInc)
s=np.std(yearInc)
new_df.loc[:,"YearlyIncomeStd"]= (new_df.loc[:,'YearlyIncome']-m) / s
new_df
```

```python
# Binarization(One Hot Encoding)

country=OneHotEncoder()
country_new_df = pd.DataFrame(country.fit_transform(new_df[['CountryRegionName']]).toarray())
country_new_df.columns=['Australia','Canada', 'United States', 'United Kingdom',
       'Germany', 'France']

# For occupation
occupation=OneHotEncoder()
occupation_new_df = pd.DataFrame(occupation.fit_transform(new_df[['Occupation']]).toarray())
occupation_new_df.columns=['Clerical', 'Skilled Manual', 'Manual', 'Management',
       'Professional']


occupation_new_df
```
